```python
# importar módulos necessários
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import math
```

# Conjunto de especificações usada para modelar os filtros

```python
# As especificações são as seguintes:
N=5 # ordem do filtro
Wn=100 # frequência crítica de stopband 
rp=5 # ripple de passband 
rs=40 # atenuação de stopband 
```

Projeto do filtro passa-baixa de Butterworth

```python
# Butterworth
b, a = signal.butter(N, Wn, 'lowpass', analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(np.abs(h)))
```

Projeto do filtro passa-baixa de Bessel

```python
# Bessel
b, a = signal.bessel(N, Wn, 'lowpass', analog=True, norm='phase')
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(np.abs(h)))
```

Projeto do filtro passa-baixa de Chebyshev Tipo I

```python
# Chebyshev Tipo I
b, a = signal.cheby1(N, rp, Wn, 'lowpass', analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))
```

Projeto do filtro passa-baixa de Chebyshev Tipo II

```python
# Chebyshev Tipo II
b, a = signal.cheby2(N, rs, Wn, 'lowpass', analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))
```

Projeto do filtro passa-baixa Elíptico

```python
# Elíptico
b, a = signal.ellip(N, rp, rs, Wn, 'lowpass', analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))
```

Comparação dos filtros de acordo com a resposta em frequência

```python
# Butterworth
b, a = signal.butter(N, Wn, 'lowpass', analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(np.abs(h)))

# Bessel
b, a = signal.bessel(N, Wn, 'lowpass', analog=True, norm='phase')
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(np.abs(h)))

# Chebyshev Tipo I
b, a = signal.cheby1(N, rp, Wn, 'lowpass', analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))

# Chebyshev Tipo II
b, a = signal.cheby2(N, rs, Wn, 'lowpass', analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))

# Elíptico
b, a = signal.ellip(N, rp, rs, Wn, 'lowpass', analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))

plt.title('Comparação de filtros passa-baixa analógicos')
plt.xlabel('Frequência [radianos / segundo]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.legend(['butter', 'bessel', 'cheby1', 'cheby2', 'ellip'])
plt.savefig("Comparacao de filtros.png",dpi=300)
plt.show()
```
